# Multiple Linear Regression Backward Elimination method

# import Librarys 

```python
# multiple linear regression # y = bo + b1x1 + b2x2
import pandas as pd
import numpy as np
from matplotlib.pyplot import plot
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.compose import ColumnTransformer
#Building the Optimal Model using Backward Elimination
import statsmodels.regression.linear_model as smf
```

```python
# importing data set
data=pd.read_csv('Startup_Data.csv')
data.head()
```

```python
# Getting the independent variables or split independent and dependent variable
real_x = data.iloc[:,0:4].values
real_y = data.iloc[:,4].values
```

# Encoding categorical data

# Encoding the Independent Variable

```python

LE = LabelEncoder()
real_x[:,3] = LE.fit_transform(real_x[:,3])
real_x[:,3]
# convert data into binary format
ct = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(categories='auto'), [3])],   # The column numbers to be transformed (here is [0] but can be [0, 1, 3])
    remainder='passthrough'                                         # Leave the rest of the columns untouched
)

real_x = ct.fit_transform(real_x)
```

# Avoiding the Dummy Variable Trap

```python
# Backword Eliminaion in we remove 1 column from preventing dummy variable trape
real_x = real_x[:,1:]
```

# Splitting the dataset into the Training set and Test set

```python
traning_x,test_x,traning_y,test_y = train_test_split(real_x,real_y,test_size=0.2,random_state=0)
```

# Fitting Multiple Linear Regression Model to the Training Set

```python
# we train the our model
MLR =  LinearRegression()
MLR.fit(traning_x,traning_y)
```

# Predicting the Test set results

```python
pred_y = MLR.predict(test_x)
pred_y
```

```python
# yo = b0  + b1x1 + b2x2 .. .. . BnXn  
'''
Adding a column in matrix of features: As we can check in our MLR equation (a), there is one constant term b0, 
but this term is not present in our matrix of features, so we need to add it manually. We will add a column
having values x0 = 1 associated with the constant term b0.
To add this, we will use append function of Numpy library(nm which we have already imported into our code), 
and will assign a value of 1. Below is the code for it.
'''
# Add columns of 1
real_x = np.append(arr=np.ones((50,1)).astype(int),values=real_x,axis=1)
real_x[:5,:]
```

```python
x_opt = real_x[:,[0,1,2,3,4,5]]
x_opt = np.array(x_opt, dtype=float)
x_opt
```

```python
# Multiple Linear Regression Model --- OLS
# train model
rel_ols = smf.OLS(endog=real_y,exog=x_opt).fit()
rel_ols.summary()
```

```python
# stil we can't get P < SL so Again
# remove first heightst value which index is 1
x_opt = real_x[:,[0,3,4,5]]
x_opt = np.array(x_opt, dtype=float)
rel_ols = smf.OLS(endog=real_y,exog=x_opt).fit()
rel_ols.summary()
```

```python
# stil we can't get P < SL so Again
# remove first heightst value which index is 4
x_opt = real_x[:,[0,3,5]]
x_opt = np.array(x_opt, dtype=float)
rel_ols = smf.OLS(endog=real_y,exog=x_opt).fit()
rel_ols.summary()
```

```python
# stil we can't get P < SL so Again
# remove first heightst value which index is 3 (0.060)
x_opt = real_x[:,[0,3]]
x_opt = np.array(x_opt, dtype=float)
rel_ols = smf.OLS(endog=real_y,exog=x_opt).fit()
rel_ols.summary()
```
